```python

import matplotlib.pyplot as plt
x = [1, 2, 3]
y = [2, 4, 6]
plt.plot(x, y, marker='o', label='Line with Markers')
plt.title('Marked Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')
plt.legend()
plt.show()
```
