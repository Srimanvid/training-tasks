```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('house-prices.csv')
df
```

```python
stat=df['Price'].describe()
print('there is a stastistics for price',stat)
sata_2=df['SqFt'].describe()
print('there is a stastistics for sqft',sata_2)
df['price_per_sqft']=df['Price'] / df['SqFt']
df.head()
```

```python
plt.scatter(df['Price'],df['SqFt'],alpha=0.5)
plt.title('house price explorer')
plt.xlabel('Price')
plt.ylabel('SqFt')
plt.show()
```
